```python
import torch
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = torch.arange(-10, 10, 0.1)
```

## Relu

```python
# max(0, x)
relu = torch.maximum(x, torch.tensor([0]))
plt.plot(x, relu)
```

## Sigmoid


```python
# 1 / (1 + e^-x)
sigmoid = 1 / (1 + torch.exp(-x))
plt.plot(x, sigmoid)
```

## Tanh


```python
# (e^x - e^-x)/(e^ + e^-x)
tanh = (torch.exp(x) - torch.exp(-x)) / (torch.exp(x) + torch.exp(-x))
plt.plot(x, tanh)
```

## Softmax

```python
# e^x/sum(e^x)
x = torch.tensor([-5.0, -2.0, 0, 20])

exp_x = torch.exp(x - x.max())

softmax = exp_x/torch.sum(exp_x)

print(softmax)
```
